```python
import openai
api_key = "Paste your key here"

openai.api_key = api_key

def get_completion(text_content):


  user_message = """I need your help generating some question answer examples based only on the information provided below.

  Please generate 10 unique examples, each containing:

  - A question about the details in the content
  - A short context/background sentence
  - The answer to the question

  It is critical that you ONLY use the details given in the content, and do not make up or add any additional information. Do not hallucinate or infer anything beyond what is provided.

  Format each example as:

  {
  "instruction": ,
  "context": ,
  "response": ,
  "category": ,
  "week": 3,
  "page": 1
  },

  The category are "closed_qa", "classification", "open_qa", "information_extraction", "brainstorming", "general_qa", "summarization", "creative_writing", based on question select the category and place it in the below format.

  Use clear, simple language focused on the key details provided. Do not add any extra fluff or length beyond what is necessary to generate the examples.

  Thank you!"""

  messages = [
    {'role':'system', 'content': text_content},
    {'role':'user', 'content': user_message}
  ]

  response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=messages,
    temperature=0,
    max_tokens=1000
  )


  summary_prompt = """Based on this content, generate a question what the text is about.

  Please generate 1 unique example, containing:

  - A question "Can you explain this slide?"
  - A short context/background sentence
  - The answer to the question

  It is critical that you ONLY use the details given in the content, and do not make up or add any additional information. Do not hallucinate or infer anything beyond what is provided.

  Format each example as:

  {
  "instruction":  ,
  "context": ,
  "response": ,
  "category": summarization ,
  "week": 3,
  "page": 1
  },

  Use clear, simple language focused on the key details provided. Do not add any extra fluff or length beyond what is necessary to generate the examples.

  Thank you!"""

  messages_summary = [
    {'role':'system', 'content': text_content},
    {'role':'user', 'content': summary_prompt}
  ]

  summary_question = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=messages_summary,
    temperature=0,
    max_tokens=1000
  )


  return summary_question.choices[0].message["content"], response.choices[0].message["content"]
```
